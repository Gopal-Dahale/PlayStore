```python
import pandas as pd
import json
import os
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
paths = []
for dirname, _, filenames in os.walk('/content/gdrive/MyDrive/PlayStore Data/All_edges_files'):
  for filename in filenames:
    paths.append(os.path.join(dirname,filename))
```

```python
# edges csv tab delimiter
c= 0
for path in paths:
  print(c,path)
  c+=1
  file = open(path)
  data= json.load(file)
  newData = list(map(lambda x: x["data"], data))
  del data
  pd.DataFrame(newData).rename(columns = {'source':'Source','target':'Target'}).to_csv('/content/gdrive/MyDrive/PlayStore Data/EdgesCSV/'+path.split('/')[-1].split('.')[0]+'.csv', sep='\t', index=False)
  del newData
  file.close()
```

```python
# nodes csv tab delimter
nodesFile = open('/content/gdrive/MyDrive/PlayStore Data/nodes.json','r',encoding='utf-8')
data = json.load(nodesFile)
newData = list(map(lambda x: x["data"], data['nodes']))
del data
pd.DataFrame(newData).rename(columns = {'id':'Id'}).to_csv('/content/gdrive/MyDrive/PlayStore Data/Nodes.csv', sep='\t', index=False)
del newData
nodesFile.close()
```

```python
# merge all edges csv
paths=  []
for dirname, _, filenames in os.walk('/content/gdrive/MyDrive/PlayStore Data/EdgesCSV'):
  for filename in filenames:
    paths.append(os.path.join(dirname,filename))
```

```python
df = pd.read_csv(paths[0],sep='\t')
for i in range(1,len(paths)):
  df = df.append(pd.read_csv(paths[i],sep='\t'))
  print(i,paths[i])
  print(len(df))
```

```python
df.to_csv('/content/gdrive/MyDrive/PlayStore Data/Edges.csv',sep='\t',index=False)
```
