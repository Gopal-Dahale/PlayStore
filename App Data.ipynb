```python
import pandas as pd
import os
```

```python
data = []
for dirname, _, filenames in os.walk('./Data'):
    for filename in filenames:
        try:
            data.append(pd.read_json(os.path.join(dirname, filename)))
        except:
            print(os.path.join(dirname, filename))
```

```python
sum = 0
for i in data:
    sum+=len(i)
```

```python
df = pd.DataFrame()
for i in data:
    df = pd.concat([df,i])
```
