```python
paths = []
for dirname, _, filenames in os.walk('/content/gdrive/MyDrive/PlayStore Data/Network_Data'):
    for filename in filenames:
        paths.append(os.path.join(dirname,filename).replace("\\","/"))
```

```python
len(paths)
```

```python
reviewers = []
c = 0
for path in paths:
    print(c,path)
    c+=1
    file = open(path)
    data = json.load(file)
    for item in data['reviewers']:
        item.update({"id":data['app']['id']})
    reviewers.extend(data['reviewers'])
    file.close()
df = pd.DataFrame(reviewers)
df.to_csv('/content/gdrive/MyDrive/PlayStore Data/Reviewers.csv',index=False)
```

```python
duplicateDf = df[df['name'].duplicated(keep=False)]
```

```python
duplicateDf.head()
```

```python
len(duplicateDf)
```

```python
grouped = duplicateDf.groupby(['name'])
```

```python
allKeys =list( grouped.groups.keys())
```

```python
lenAllKeys = len(allKeys)
lenAllKeys 
```

```python
def edgesBasic(start,end):
    c = start
    for person in allKeys[start:end]:
        file = open('./Edges_Basic/' +str(c)+ '.json','w',encoding="utf-8")
        print(c,person)
        c+=1
        obj = {person : duplicateDf.loc[grouped.groups[person]][['score','id']].to_dict(orient='records')}
        jsonObj = json.dumps(obj,indent = 2)
        file.write(jsonObj)
        file.close()
```
