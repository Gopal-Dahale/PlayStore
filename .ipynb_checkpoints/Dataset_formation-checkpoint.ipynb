```python
import pandas as pd
import os
```

```python
# order of categories list matters
categories = ["APPLICATION","ANDROID_WEAR","ART_AND_DESIGN","AUTO_AND_VEHICLES","BEAUTY","BOOKS_AND_REFERENCE",
              "BUSINESS","COMICS","COMMUNICATION","DATING","ENTERTAINMENT","EVENTS","FINANCE",
              "FOOD_AND_DRINK","HEALTH_AND_FITNESS","HOUSE_AND_HOME","LIBRARIES_AND_DEMO","LIFESTYLE",
              "MAPS_AND_NAVIGATION","MEDICAL","MUSIC_AND_AUDIO","NEWS_AND_MAGAZINES","PARENTING","PERSONALIZATION",
              "PHOTOGRAPHY","PRODUCTIVITY","SHOPPING","SOCIAL","TOOLS","TRAVEL_AND_LOCAL","VIDEO_PLAYERS",
              "WEATHER","FAMILY_ACTION","FAMILY_BRAINGAMES","FAMILY_CREATE","FAMILY_EDUCATION","FAMILY_MUSICVIDEO",
              "FAMILY_PRETEND","FAMILY","GAME_ACTION","GAME_ADVENTURE","GAME_ARCADE","GAME_BOARD","GAME_CARD",
              "GAME_CASINO","GAME_CASUAL","GAME_EDUCATIONAL","GAME_MUSIC","GAME_PUZZLE","GAME_RACING","GAME_ROLE_PLAYING",
              "GAME_SIMULATION","GAME_SPORTS","GAME_STRATEGY","GAME_TRIVIA","GAME_WORD","EDUCATION","GAME","SPORTS",
              "GOOGLE_CAST","KIDS"]

collections = ["TOP_FREE","TOP_PAID","GROSSING","TRENDING", "NEW_FREE","NEW_PAID"]
```

```python
data = []
for dirname, _, filenames in os.walk('./Data'):
    for filename in filenames:
        fileName = os.path.join(dirname, filename).split('\\')[-1] # Get the file name
        fileCat = ""
        fileColln = ""
        for cat in categories:
            if(cat in fileName):
                fileCat = cat
                break
        for colln in collections:
            if(colln in fileName):
                fileColln = colln
                break
        
        if(fileColln == ""):
            for colln in ["FREE", "PAID"]:
                if(colln in fileName):
                    fileColln = "TOP_" + colln
        
        df = pd.read_json(os.path.join(dirname, filename)) # read file
        df['category'] = fileCat
        df['grossing'] = 0
        df['trending'] = 0
            
        if(fileColln == 'GROSSING'):
            df['grossing'] = 1
        elif(fileColln == 'TRENDING'):
            df['trending'] = 1
        
        data.append(df.copy())
```

```python
len(data)
```

```python
sum = 0
for i in data:
    sum+=len(i)
print(sum)
```

```python
df = pd.DataFrame()
for i in data:
    df = pd.concat([df,i])
```

```python
df.drop_duplicates(subset="appId",inplace=True)
```

```python
df.to_csv('Dataset.csv',index=False)
```
