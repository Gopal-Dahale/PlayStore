```python
import pandas as pd
import os
import json
```

```python
paths = []
for dirname, _, filenames in os.walk('./Reviews'):
    for filename in filenames:
        paths.append(os.path.join(dirname,filename).replace("\\","/"))
```

```python
len(paths)
```

```python
data = pd.read_csv('./Dataset.csv')
```

```python
# create subdirectories in Network_Data Folder
dirNames = list(data['category'].drop_duplicates())
path = './Network_Data/'
for dirName in dirNames:
    os.mkdir(path+str(dirName))
```

```python
app = {}
def networkData(start,end):
    count = start
    for path in paths[start:end]:
        appId = path.split('/')[-1][0:-4]
        appDetails = data.loc[data['appId'] == appId].values[0]
        print(count,appId)
        count+=1
        app['id'] = appId
        app['price'] = appDetails[7] # price
        app['rating'] = appDetails[10] # scoreText
        app['category']= appDetails[13] # category
        app['grossing']= appDetails[14] # grossing
        app['trending']= appDetails[15] # trending
        app['icon'] =  appDetails[3] # icon 
        
        try:
            df = pd.read_csv(path) # reviews file for the app
            df.drop_duplicates(subset = "userName",inplace = True) # remove duplicate user in same reviews file of an app
            names = list(map(lambda item: {"name":item},df['userName'].tolist())) 
            scores = list(map(lambda item: {'score': item},df['score'].tolist())) 
            for i,j in zip(names,scores):
                i.update(j)
            obj = {"app":app,"reviewers": names}    
        except:
            obj = {"app":app,"reviewers":[]}
            print("The csv file contains no rows");
            
        jsonObj = json.dumps(obj, indent = 2)
        file = open('./Network_Data/'+ app['category'] + '/' + appId+'.json','w')
        file.write(jsonObj)
        file.close()
```

```python
networkData(0,57583)
```
