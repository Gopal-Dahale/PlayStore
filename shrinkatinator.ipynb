```python
G = nx.from_pandas_edgelist(edges,'Source',"Target",["interaction","source-score","target-score"]) # create graph
```

```python
del edges
```

```python
print(nx.info(G))
```

```python
# sampling
n = G.number_of_nodes()
print("Number of nodes in original graph:", n)
sub_n = n//4
print("Number of nodes in Sampled graph:" , sub_n)
index_of_first_random_node = random.randint(0, n - 1)
print("Index of first random node:",index_of_first_random_node)
nodes_list = list(list(G.nodes()))
sampled_nodes = set([nodes_list[index_of_first_random_node]])
```

```python
iteration = 1
nodes_before_t_iter = 0
curr_node = nodes_list[index_of_first_random_node]
T = 100 # number of iterations
growth_size = 2

while len(sampled_nodes) != sub_n:
  edges = [n for n in G.neighbors(curr_node)]
  index_of_edge = random.randint(0, len(edges) - 1)
  chosen_node = edges[index_of_edge]
  sampled_nodes.add(chosen_node)
  curr_node = chosen_node
  iteration = iteration + 1
  print(len(sampled_nodes))
  
  if iteration % T == 0:
    if ((len(sampled_nodes) - nodes_before_t_iter) < growth_size):
      curr_node = nodes_list[random.randint(0, n - 1)]
    nodes_before_t_iter = len(sampled_nodes)

  del edges
```

```python
H = G.subgraph(sampled_nodes)
```

```python
print(nx.info(H))
```

```python
sampledEdges = nx.to_pandas_edgelist(H,source = "Source", target = "Target")
```

```python
sampledEdges.to_csv('/content/gdrive/MyDrive/PlayStore Data/Sampled Graph/SampledEdges.csv',encoding='utf-8',sep = '\t',index =False)
```
