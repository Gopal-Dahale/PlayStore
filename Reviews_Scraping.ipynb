```python
import pandas as pd
from google_play_scraper import Sort, reviews
import time
```

```python
df = pd.read_csv('./Dataset.csv')
```

```python
total = len(df)-1
total 
```

```python
def FetchReviews(start,end):
    count = start
    f = open("./went_wrong.txt","a")
    for appId in df['appId'][start:end]:
        token = None
        data = []
        print(count,"/",total, appId)
        for i in range(5):
            print('Round',i+1,end=' ')
            try:
                res,t =  reviews(appId, lang='en',country='in',sort=Sort.MOST_RELEVANT,count=200,continuation_token=token)
                time.sleep(1)
            except:
                print('Something went wrong with', appId)
                f.write(appId+"\n")
                break
            token = t
            data += res
        count += 1
        print('Reviews Count', len(data))
        pd.DataFrame(data).to_csv("./Reviews/"+ appId +".csv",index=False)
    f.close()
```
